### DATA MANIPULATION

```python
# no. of missing values in each column.
titanic_data.isnull().sum()
```

```python
# as we don't need the cabin column, i am dropping it.
titanic_data = titanic_data.drop(columns='Cabin', axis=1)
```

```python
# filling missing values in age column with mean value.
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)
```

```python
# let's see which value is being written more which is called mode value.
print(titanic_data['Embarked'].mode())
# this implies that most people were from southampton.
```

```python
# replacing missing values in embarked with mode value which is from index [0].
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
```

```python
# now, all the values have been filled, no column is empty.
titanic_data.isnull().sum()
```

Encoding the Categorical Columns.

```python
titanic_data['Sex'].value_counts()
```

```python
titanic_data['Embarked'].value_counts()
```

```python
titanic_data.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
titanic_data.head()
```

Seperating features and Target.

```python
X = titanic_data.drop(columns = ['PassengerId', 'Name', 'Ticket', 'Survived'], axis=1)
Y = titanic_data['Survived']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training and Test Data.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)  # 20% data is being tested(0.2), random state is used to split the data in a formal way or in the same way.
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

```python
train = model.predict(X_train)
print(train)
```

```python
TrainingAccuracy = accuracy_score(Y_train, train)
print('Accuracy score of the training data is : ', TrainingAccuracy)
```

```python
test = model.predict(X_test)
print(test)
```

```python
TestAccuracy = accuracy_score(Y_test, test)
print('Accuracy of tested data is : ', TestAccuracy)
```
